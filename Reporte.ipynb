## CRC-32

```python
ceros_eCRC = eCRC["w_noise"].value_counts()
ceros_rCRC = rCRC["err"].value_counts()
print("Cantidad de cadenas correcta(0) y modificadas(1) enviadas")
print(ceros_eCRC)
print("Cantidad de cadenas correcta(0) y modificadas(1) recibidas")
print(ceros_rCRC)
```

## Hamming

```python
# Buscando si todas las tramas transferidas y corregidas son exáctamente las mismas

cont = 0
for x in range(len(eHam)):
    if eHam.trama_og[x] != rHam.orig[x]:
        cont += 1
        
print("Cantidad de tramas que difieren entre emisor y receptor: ", cont)
```
