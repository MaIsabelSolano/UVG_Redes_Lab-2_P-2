## CRC-32

```python
ceros_eCRC = eCRC["w_noise"].value_counts()
ceros_rCRC = rCRC["err"].value_counts()
print("Cantidad de cadenas correcta(0) y modificadas(1) enviadas")
print(ceros_eCRC)
print("Cantidad de cadenas correcta(0) y modificadas(1) recibidas")
print(ceros_rCRC)
```

```python
# Data for the pie chart
labels = ['correctas', 'modificadas']
sizes = [ceros_eCRC[0], ceros_eCRC[1]]  # Percentages or values for each category

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title("Cadenas correctas e incorrectas, CRC")

# Show the pie chart
plt.show()
```

```python
ceros_eHAM = eHam["w_noise"].value_counts()
print("Cantidad de cadenas correcta(0) y modificadas(1) enviadas")
print(ceros_eHAM)
```

```python
switches_general = 0
digitos_general = 0
porPalabra = []

for x in eCRC['id']:
    # for y in eCRC['codified'][x][1:-1]:
    for i in range(len(eCRC['codified'][x])):
        if eCRC['codified'][x] != "'":
            digitos_general += 1
            
            if eCRC['codified'][x][i] != eCRC['modified'][x][i]:
                switches_general += 1
                        
print(switches_general)
print(digitos_general)
```

```python
# Data for the pie chart
labels = ['switches', 'Ã­ntegros']
sizes = [switches_general, digitos_general - switches_general]  # Percentages or values for each category

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title("Porcentajes de bits cambiados e Ã­ntegros, CRC")

# Show the pie chart
plt.show()
```

## Hamming

```python
# Buscando si todas las tramas transferidas y corregidas son exÃ¡ctamente las mismas

cont = 0
for x in range(len(eHam)):
    if eHam.trama_og[x] != rHam.orig[x]:
        cont += 1
        
print("Cantidad de tramas que difieren entre emisor y receptor: ", cont)
```

```python
# Data for the pie chart
labels = ['difieren', 'Correctos']
sizes = [cont, 500-cont]  # Percentages or values for each category

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title("Tramas que difieren entre emisor y receptor, Hamming")

# Show the pie chart
plt.show()
```
